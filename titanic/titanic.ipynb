Import libraies:

```python
import pandas as pd
import numpy as np
```

```python
def process_data(data):
    # drop irrelevant data
    data = data.drop(['PassengerId','Name','Ticket','Cabin'], axis=1)
    # convert to numeric format
    data['Sex'] = data['Sex'].map({'male':1, 'female':0})
    data['Embarked'] = data['Embarked'].map({'S':0, 'C':1, 'Q':2})
    # clean nan age 
    mean_age = data['Age'].mean()
    age_nan_to_mean = lambda x: mean_age if np.isnan(x) else x
    data['Age'] = data['Age'].map(age_nan_to_mean)
    
#     data = data.drop(['Age', 'Fare'], axis=1)
    return data
```

```python
def feature_label_split(data):
#     print(data.loc[:, 'Pclass':].values[:2])
    return data.loc[:, 'Pclass':].values, data.loc[:, :'Survived'].values
```

```python
data = pd.read_csv("train.csv")
data = process_data(data)
data = data.dropna()
print('null counts ', data.isnull().sum(axis=0))

train_x, train_y = feature_label_split(data)
# train_x = train_x.reshape(train_x.shape[0], train_x.shape[1])
train_y = train_y.reshape(train_y.shape[0], )
print('x shape:', train_x.shape)
# print(train_x[:5])
print('y_shape:', train_y.shape)
# print(train_y[0])
```

```python
x = train_x[:20]
y = train_y[:20]
print(x.shape)
x = x.reshape(x.shape[0],x.shape[1])
print(x.shape)
print(x)
print(y)
import keras
from keras.models import Sequential
from keras.layers import  Dense, Activation
model = Sequential()
model.add(Dense(3, input_dim=7))
model.add(Activation('sigmoid'))
model.add(Dense(1))
model.add(Activation('sigmoid'))
model.compile(optimizer='adam',
             loss=keras.losses.binary_crossentropy,
             metrics=['accuracy'])
model.fit(x, y, epochs=1000)
print('real result = ',train_y,' result is ',model.predict(x))
```

Construct nn model:

```python
import keras
from keras.models import Sequential
from keras.layers import  Dense, Dropout, Flatten, Activation, BatchNormalization, regularizers
```

```python
def nn_model(input_dim):
    model = Sequential()
#     model.add(BatchNormalization(input_shape=(input_dim,), name="input_batch_norm"))
    model.add(Dense(32, input_dim = input_dim, name='l1'))
    model.add(Activation('sigmoid', name='a1'))
    model.add(Dense(16, name='l2'))
    model.add(Activation('relu', name='a2'))
    model.add(Dense(1, name='lo'))
    model.add(Activation('sigmoid', name='ao'))
    
    return model
```

```python
# create model
print(train_x.shape)
model = nn_model(train_x.shape[1])
model.compile(optimizer='adam',
             loss=keras.losses.binary_crossentropy,
             metrics=['accuracy'])

# train model
# print(train_x[61])
# print(train_x[0])
# print(train_y)
model.fit(train_x, train_y, epochs=250, batch_size=32)
# model.evaluate(train_x, train_y)
```
